# <font color=#4267b2>擊敗老師的神經網路 (MNIST手寫數字辨識)</font>
## <font color=#4267b2>設定及匯入需要的套件</font>

```python
%env KERAS_BACKEND=tensorflow
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mrl
mpl.rc("font", family = "Noto Sans CJK TC")
from keras.datasets import mnist
from ipywidgets import interact_manual
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD, RMSprop, Adagrad, Adadelta, Adam, Adamax, Nadam
from keras.optimizers import Adadelta
```

## <font color=#4267b2>載入MNIST資料<font>

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

### <font color=#4267b2>將測試資料 x 除以最大值(255)<font>  
### <font color=#4267b2>將測試資料 y 進行 one-hot 處理<font>

```python
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
x_train_n = x_train / 255. 
x_test_n = x_test/ 255.   
y_train = np_utils.to_categorical(y_train, num_classes=10)
y_test = np_utils.to_categorical(y_test, num_classes=10)
```

### <font color=#4267b2>改變不同optimizer(SGD以及其他)<font>
第一、二層神經元固定不變各爲4,2  
第一、二層Activation 函數仍使用 `sigmoid`  
第三層使用 `softmax`  
這裏只爲了比較各函數差距  
所以使用batch_size=10, epochs=15，減少計算時間

```python
model_list= []
model_fit_list= []
model_name = ['SGD', 'RMSprop', 'Adagrad', 'Adadelta','Adam', 'Adamax', 'Nadam']
model_optimizer = [SGD(lr=0.087),RMSprop(lr=0.01, rho=0.9, epsilon=1e-06),
                   Adagrad(lr=0.001, epsilon=1e-06),Adadelta(lr=3, rho=0.95, epsilon=2e-06),
                  Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08),Adamax(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08),
                  Nadam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, schedule_decay=0.004)]

for i in range(7):
    model = Sequential()
    model.add(Dense(4, input_dim=784))
    model.add(Activation('sigmoid'))
    model.add(Dense(2))
    model.add(Activation('sigmoid'))
    model.add(Dense(10))
    model.add(Activation('softmax'))
    model.compile(loss='mse', optimizer=model_optimizer[i], metrics=['accuracy'])
    model_fit=model.fit(x_train, y_train, batch_size=10, epochs=15,verbose = 0,validation_data = (x_test,y_test))
    model_list.append(model)
    model_fit_list.append(model_fit)
```

```python
plt.figure(dpi=100)
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
for i in range(7):
    plt.plot(model_fit_list[i].history["acc"], label=model_name[i])
plt.legend(loc=(1.01,0.43))
plt.tight_layout(pad=0.4, w_pad=1.2, h_pad=1.0)
```

## <font color=#4267b2>嘗試使用不同激發函數<font>
在下方測試後，第一層使用`selu`、第二層使用`relu`  
optimizer 使用 `Nadam`  
loss function 使用 `categorical_crossentropy` 原本就是用來做多種類別分類  
測試資料 x 使用正規後的結果來訓練  

```python
model_list_relu= []
model_fit_list_relu= []
model_name = 'Nadam'
model_optimizer = Nadam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, schedule_decay=0.004)

for i in range(1):
    model = Sequential()
    model.add(Dense(4, input_dim=784))
    model.add(Activation('selu'))
    model.add(Dense(2))
    model.add(Activation('relu'))
    model.add(Dense(10))
    model.add(Activation('softmax'))
    model.compile(loss='categorical_crossentropy', optimizer=model_optimizer, metrics=['accuracy'])
    model_fit=model.fit(x_train_n, y_train, batch_size=10, epochs=15,verbose = 0,validation_data = (x_test_n,y_test))
    model_list_relu.append(model)
    model_fit_list_relu.append(model_fit)
plt.figure(dpi=100)
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
for i in range(1):
    plt.plot(model_fit_list_relu[i].history["acc"], label=model_name)
plt.legend(loc=(1.01,0.43))
plt.tight_layout(pad=0.4, w_pad=1.2, h_pad=1.0)
```

### <font color=#4267b2>使用不同數量的神經元<font>
一開始想堅持使用第一層4個、第二層2的神經元的數量調變學習率  
但後來發現不管怎麼調整都很難突破準確率 0.8  
因此目標改成"**<font color=#ef6c00>利用最少神經元組合數使超過準確率 0.9</font>**"  
層數仍是兩層但第一層有(2,10)，9個值  
第二層有(2,6)，5個值

```python
model_name = 'Nadam'
model_optimizer = Nadam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, schedule_decay=0.009)
x, y = np.meshgrid(np.linspace(2,10,9),np.linspace(2,6,5))
model_acc = np.zeros(45)
model_acc = model_acc.reshape(5,9)
for j in range(9):
    for i in range(5):
        model = Sequential()
        model.add(Dense(int(x[i][j]), input_dim=784))
        model.add(Activation('selu'))
        model.add(Dense(int(y[i][j])))
        model.add(Activation('relu'))
        model.add(Dense(10))
        model.add(Activation('softmax'))
        model.compile(loss='categorical_crossentropy', optimizer=model_optimizer, metrics=['accuracy'])
        model_fit=model.fit(x_train_n, y_train, batch_size=10, epochs=10,verbose = 0,validation_data = (x_test_n,y_test))
        model_acc[i][j] = model_fit.history["acc"][-1]
```

```python
x, y = np.meshgrid(np.linspace(2,10,9),np.linspace(2,6,5))
plt.figure(dpi=100)
plt.imshow(model_acc, origin=[0,0],extent=(1.5,10.5,1.5,6.5))
plt.colorbar()
```

### <font color=#4267b2>畫出準確率0.9達標的點</font>
並畫出兩層神經元相乘等高線

```python
x, y = np.meshgrid(np.linspace(2,10,9),np.linspace(2,6,5))
plt.scatter(x[model_acc>0.9],y[model_acc>0.9])
x_1 = np.linspace(6,10)
y_val = [24, 30, 32, 35,36]
for y_v in y_val:
    plt.plot(x_1,y_v/x_1)
```
