```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#import the data from "data" folder
result1 = "data/test_05_9GHz.txt"
data1 = pd.read_csv(result1,sep="\t")
result2 = "data/test_06_7GHz.txt"
data2 = pd.read_csv(result2,sep="\t")
result3 = "data/test_07_10GHz.txt"
data3 = pd.read_csv(result3,sep="\t")
result4 = "data/test_08_11GHz.txt"
data4 = pd.read_csv(result4,sep="\t")
```

```python
%matplotlib inline
# remove the first spike in my data
x1 = data1["Field (G)"][1:]
y1 = data1["S12_Log Mag. (dB)"][1:] - data1["S12_Log Mag. (dB)"][1]
x2 = data2["Field (G)"][1:]
y2 = data2["S12_Log Mag. (dB)"][1:] - data2["S12_Log Mag. (dB)"][1]
x3 = data2["Field (G)"][1:]
y3 = data3["S12_Log Mag. (dB)"][1:] - data3["S12_Log Mag. (dB)"][1]
x4 = data4["Field (G)"][1:]
y4 = data4["S12_Log Mag. (dB)"][1:] - data4["S12_Log Mag. (dB)"][1]

# plotting and modifying figure
plt.figure(dpi=1000)
line1 = plt.plot(x2, y2, color="black")
line2 = plt.plot(x1,y1, color="red")
line3 = plt.plot(x3,y3, color="blue")
line4 = plt.plot(x4,y4, color="green")
legend = plt.legend(["7  GHz","9  GHz","10GHz","11GHz"])
plt.ylim(-0.1, 0.027)
plt.xlim(-300,2700)
plt.show()
```
